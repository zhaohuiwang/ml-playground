```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
wine_data = datasets.load_wine(as_frame=True)
df = wine_data.data
print(df.shape)
df
```

```python
std_scaler = StandardScaler()
scaled_df = std_scaler.fit_transform(df)
scaled_df
```

```python

pca = PCA(n_components=3)
pca.fit_transform(scaled_df)
```

```python
print(pca.components_)
print(sum(pca.explained_variance_ratio_))
```

```python
pca = PCA()
pca.fit(scaled_df)
exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)
exp_var_cumul = np.insert(exp_var_cumul, 0, [0])
```

```python

plt.figure(figsize=(4,2),dpi=150)
plt.grid()
plt.plot(range(len(exp_var_cumul)), exp_var_cumul,  marker='o')
plt.xlabel('n_components')
plt.ylabel('Explained variance ratio')
plt.title('n_components vs. Explained Variance Ratio')
plt.show()
```

### The Magic of Principal Component Analysis through Image Compression

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import numpy as np
from sklearn.decomposition import PCA, IncrementalPCA

pca = PCA()
```

```python
image_raw = imread("../../data/images/dragonfly.jpg")
print(image_raw.shape)

plt.figure(figsize=[12,8])
plt.imshow(image_raw)
```

```python
# Show the new shape of the image
print(f"the raw image dimension: \t{image_raw.shape}")
image_sum = image_raw.sum(axis=2)
# sum along the third axis with RGB values. The result is one dimension less.
print(f"the modified image dimension: \t{image_sum.shape}")

# rescale the value (min-max rescale). 0.0 = White, 1.0 = Black.
image_bw = (image_sum-image_sum.min())/(image_sum.max()--image_sum.min())
print(f"Value range \t{image_bw.min():.4f} - {image_bw.max():.4f}")
```

```python

pca.fit(image_bw)

# Getting the cumulative variance
var_cumu = np.cumsum(pca.explained_variance_ratio_)*100

# How many PCs explain 95% of the variance?
k = np.argmax(var_cumu>95)
print("Number of components explaining 95% variance: "+ str(k))
#print("\n")

plt.figure(figsize=[10,5])
plt.title('Cumulative Explained Variance explained by component')
plt.ylabel('Cumulative Explained variance (%)')
plt.xlabel('Principal components')
plt.axvline(x=k, color="k", linestyle="--")
plt.axhline(y=95, color="r", linestyle="--")
ax = plt.plot(var_cumu)
```

```python
# reduce the dimensionality with PCA and compare the image of selected dimensions
def plot_at_k(k):
    ipca = IncrementalPCA(n_components=k)
    # IncrementalPCA divides data into batches instead to perform PCA, for memory efficiency
    image_recon = ipca.inverse_transform(ipca.fit_transform(image_bw))
    plt.imshow(image_recon,cmap = plt.cm.gray)

ks = [10, 25, 50, 100, 150, 250]

plt.figure(figsize=[15,9])

for i in range(len(ks)):
    plt.subplot(2,3,i+1)
    plot_at_k(ks[i])
    plt.title("Components: "+str(ks[i]))

plt.subplots_adjust(wspace=0.2, hspace=0.0)
plt.show()
```
